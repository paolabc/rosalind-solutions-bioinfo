```python
# The offspring rabbit needs one month to grow and then they could have k pars of offspring.

# A Fibonacci sequence, we need to take account the rabbits pair in the month before
# and the present month to get the answer

# 5 MONTH AND 2 OFFSPRING PAIRS
# Month 1 -> p 
# Month 2 -> P 
# Month 3 -> P p p 
# Month 4 -> P p p  P P 
# Month 5 -> P p p  P P  P p p  P p p 
 
# 5 MONTH AND 3 OFFSPRING PAIRS
# Month 1 -> p 
# Month 2 -> P 
# Month 3 -> P p p p 
# Month 4 -> P p p p P P P 
# Month 5 -> P p p p P P P P p p p P p p p P p p p

#Fn = Fn-1 + Fn-2 * k , where k is the number of offspring pars in each cycle

#sample
month,number_of_pars= (35,5)


#Solution
count=1
#the fisrt month 
child=1
#the second month
rabbit_pars=1 #the number of rabbit pairs that could mate


for m in range(1,month-1):
        keep_rabbit_pars=rabbit_pars
        rabbit_pars= rabbit_pars+child*number_of_pars
        child= keep_rabbit_pars
        
      
print(rabbit_pars)      
     
```
